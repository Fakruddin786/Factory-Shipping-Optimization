# Exploratory Data Analysis
This notebook demonstrates data cleaning and initial EDA.

```python
import pandas as pd
import numpy as np
import plotly.express as px
import os

data_path = os.path.join('..', 'data', 'orders.csv')
df = pd.read_csv(data_path)
df.head()
```

## Data Cleaning

```python
print(f'Original shape: {df.shape}')
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
print(f'Clean shape: {df.shape}')
```

## Visualizations

```python
fig = px.histogram(df, x='Lead_Time_Days', title='Distribution of Lead Times')
fig.show()
```

```python
corr = df[['Distance_KM', 'Sales', 'Gross_Profit', 'Lead_Time_Days', 'Profit_Margin']].corr()
fig2 = px.imshow(corr, title='Correlation Heatmap', text_auto=True)
fig2.show()
```
